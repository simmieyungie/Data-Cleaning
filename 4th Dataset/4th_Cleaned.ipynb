```python
import pandas as pd
import numpy as np
```

```python
Cdata = pd.read_excel("4.-Badly-Structured-Sales-Data-4.xlsx")
```

```python
#Checking the Data structure
Cdata
```

```python
Cdata.columns
```

```python
#Renaming the Columns
Cdata.rename(columns = {'Unnamed: 0': 'Order ID','Ship Mode':'Order Date' }, inplace = True)
Cdata.rename(columns= {'Unnamed: 3':'First Class', 'Unnamed: 4':'First Class'}, inplace = True)
Cdata.rename(columns= {'Unnamed: 6':'Same Day', 'Unnamed: 7': 'Same Day' }, inplace= True)
Cdata.rename(columns = {'Unnamed: 9':'Second Class', 'Unnamed: 10':'Second Class' }, inplace = True)
Cdata.rename(columns= {'Unnamed: 12':'Standard Class', 'Unnamed: 13':'Standard Class'}, inplace = True)
```

```python
#Extracting the data to edit the first part (Clean Data 1 = Cdata1)
Cdata1 = Cdata.iloc[2:-1]
```

```python
Cdata1
```

```python
#Resetting the Index 
Cdata1.reset_index(drop= True, inplace = True)
```

```python
#Turning the Coulunms into rows 
Cdata1= pd.melt(Cdata1, id_vars= ['Order ID', 'Order Date'], value_vars= ['First Class', 'First Class', 'First Class',
                                        'Same Day', 'Same Day', 'Same Day', 'Second Class', 'Second Class',
                                        'Second Class', 'Standard Class', 'Standard Class', 'Standard Class'],
                                            var_name='Ship Mode', value_name='Sales')
```

```python
Cdata1
```

```python
#Extracting the data to edit the Second Phase (Clean Data 2 = Cdata2)
Cdata2 = Cdata.iloc[0:-1]
```

```python
Cdata2
```

```python
#Making the First Row the New Column
Cdata2.columns = Cdata2.iloc[0]
Cdata2.columns
```

```python
#Renaming the column
Cdata2.rename(columns = {np.nan:'Order ID'}, inplace = True)
```

```python
Cdata2.rename (columns = {'Segment': 'Order Date'}, inplace = True)
```

```python
Cdata2.columns
```

```python
#Dropping unwanted rows
Cdata2.drop(0, inplace = True)
Cdata2.drop(1, inplace = True)
```

```python
#Resetting the index
Cdata2.reset_index(drop= True, inplace = True)
```

```python
##Turning the Coulunms into rows 
Cdata2 = pd.melt(Cdata2, id_vars= ['Order ID', 'Order Date'], value_vars= ['Consumer', 'Corporate', 'Home Office',
                                                   'Consumer', 'Corporate', 'Home Office', 'Consumer', 'Corporate',
                                                    'Home Office', 'Consumer', 'Corporate', 'Home Office'],
                                                    var_name='Segment', value_name='Sales' )
```

```python
Cdata2
```

```python
#merging the two cleaned phases
full = pd.merge(Cdata1, Cdata2, on = ('Order ID', 'Order Date', 'Sales'))
```

```python
#dropping Null values
full.dropna(inplace = True )
```

```python
#Dropping the duplicates but keeping the first
full.drop_duplicates(subset ="Order ID",
                     keep = 'first', inplace = True) 
```

```python
#resetting the index
full.reset_index(drop= True, inplace = True)
```

```python
full
```

```python
#saving the cleaned data as a csv file
full.to_csv('file_name2.csv', index=False)
```
